```python
def to_midpoint(range_str):
    """Convert a string representation of a range to its midpoint."""
    try:
        start, end = map(int, range_str.split(' - '))
        return (start + end) // 2
    except ValueError:
        return np.nan
```

```python
import ast

def clean_text_column(df, columns):
    """Clean and normalize text columns for ML usage."""
    for column in columns:
        if column == 'detailed_description':
            df[column] = df[column].astype(str).str.lower()
            continue

        if column == 'tags':
            df[column] = df[column].apply(
                lambda x: list(ast.literal_eval(x).keys()) if isinstance(x, str) and x != '[]' else []
            )
        else:
            # Convert stringified lists to actual lists
            df[column] = df[column].apply(
                lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else []
            )

        # Normalize text: lowercase, replace spaces, join into one string
        df[column] = df[column].apply(
            lambda lst: ' '.join(i.lower().replace(' ', '_') for i in lst) if isinstance(lst, list) else ''
        )

    return df
```

# Recommendation model

```python
import pandas as pd
import numpy as np
import os
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import hstack
from sklearn.neighbors import NearestNeighbors

df = pd.read_parquet(os.path.join('data', 'processed', 'games_march2025_cleaned_ready_ml.parquet'))
df.head(1)
```

```python
df.info()
```

```python
sorted(df.loc[df['name'].str.contains('ELDEN', na=False)].name)[:50]
```

```python
# Create TF-IDF matrix for detailed_description
tfidf_desc = TfidfVectorizer(stop_words='english')
desc_matrix = tfidf_desc.fit_transform(df['detailed_description'])
desc_matrix
```

```python
# Create TF-IDF matrix for tags_combined
tfidf_tags = TfidfVectorizer()
tags_matrix = tfidf_tags.fit_transform(df['tags_combined'])
tags_matrix
```

```python
game = df.loc[df['name'].str.contains('ELDEN', na=False)].copy(deep=True)
game
```

```python
df.loc[df['name'].str.contains('NIOH', na=False, case=False)]
```

```python
game_elden_ring = game
game_dark_souls = df.loc[df['name'] == 'DARK SOULS™ III'].copy(deep=True)
game_nioh = df.loc[df['name'] == 'Nioh: Complete Edition'].copy(deep=True)
games = pd.concat([game_elden_ring, game_dark_souls, game_nioh], ignore_index=True)
games
```

```python
games_most_popular_platform = pd.DataFrame(games['platforms'].tolist()).sum().idxmax()
games_most_popular_platform
```

```python
games = df.loc[df['name'].isin(['ELDEN RING', 'DARK SOULS™ III', 'Nioh: Complete Edition'])].copy(deep=True).index
games
```

```python
weights = {
    # Similarity
    "tags": np.sqrt(0.45),
    "desc": np.sqrt(0.45),
    "game_age": np.sqrt(0.1),
    # After nearest neighbors
    "user_attractiveness": 0.25,
    'platforms': 0.1,
}
```

```python
game_age_matrix = df[['game_age']].values
game_age_matrix
```

## Nearest Neighbors on Combined Vector

```python
# Apply weights
desc_matrix = desc_matrix.multiply(weights['desc'])
tags_matrix = tags_matrix.multiply(weights['tags'])
game_age_matrix_weighted = game_age_matrix * weights['game_age']

# Combine all into one matrix
combined_matrix = hstack([
    tags_matrix,
    desc_matrix,
    game_age_matrix_weighted,
])
combined_matrix = combined_matrix.tocsr()
```

```python
# Fit Nearest Neighbors
nn_model = NearestNeighbors(metric='cosine')
nn_model.fit(combined_matrix)
```

```python
# Get the games we want to find similar games for
elden_idx = df.loc[df['name'].str.contains('ELDEN', case=False, na=False)].index[0]
elden_vector = combined_matrix[elden_idx]
```

```python
games_indxs = games
games_indxs
```

```python
elden_vector = combined_matrix[games_indxs]
```

```python
# Find similar games (30 + possibly the game itself)
distances, indices = nn_model.kneighbors(elden_vector, n_neighbors=30 + elden_vector.shape[0])
indices = indices[0]
distances = distances[0]
```

```python
type(indices)
```

```python
# Remove the games itself from the results
mask = ~np.isin(indices, games_indxs)
filtered_indices = indices[mask]
filtered_distances = distances[mask]
```

```python
# Compute cosine similarity scores (Cosine similarity: 1 - distance)
base_similarities = 1 - filtered_distances
```

```python
# Boost similarities based on user_attractiveness
user_attractiveness = df.loc[filtered_indices, 'user_attractiveness'].values
base_similarities += user_attractiveness * weights['user_attractiveness']
```

```python
# Boost similarities based on platforms
platforms = df.loc[filtered_indices, 'platforms'].values
base_similarities *= np.array([1 + weights['platforms'] if i[games_most_popular_platform] else 1. for i in platforms])
```

```python
1 - filtered_distances
```

```python
base_similarities
```

```python
# Sort by similarity score
indices_scores = np.array(sorted(zip(filtered_indices, base_similarities), key=lambda x: x[1], reverse=True))

# Limit to the top 10 recommendations
top_n = 10
top_indices = indices_scores[:top_n]
top_indices
```

```python
# Scale the final similarities to be between 0 and 1
top_indices[:, 1] = np.clip(top_indices[:, 1], 0, 1)
top_indices
```

```python
# Retrieve the recommendations from the DataFrame and add similarity scores
results = df.iloc[top_indices[:, 0]].copy()
results['similarity'] = top_indices[:, 1]

print("Top game recommendations similar to Elden Ring:")
print(results[['name', 'similarity']])
```
